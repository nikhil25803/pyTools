```python
import random
import os
import shutil
from io import BytesIO
from PyPDF2 import PdfReader, PdfWriter
import mimetypes
```

```python
import PyPDF2

# Set the new dimensions of the PDF pages
new_width = 1920  
new_height = 1080  

# Open the input PDF file
with open('./debu.pdf', 'rb') as file:
    # Read the PDF file
    pdf_reader = PyPDF2.PdfReader(file)

    # Create a new PDF writer object
    pdf_writer = PyPDF2.PdfWriter()

    # Loop through the pages of the input PDF file
    for page_num in range(pdf_reader.numPages):
        # Get the current page
        page = pdf_reader.getPage(page_num)

        # Resize the page
        page.mediaBox.lowerLeft = (0, 0)
        page.mediaBox.upperRight = (new_width, new_height)

        # Add the resized page to the PDF writer object
        pdf_writer.addPage(page)

    # Open a new output PDF file
    with open('output.pdf', 'wb') as output_file:
        # Write the output PDF file
        pdf_writer.write(output_file)
```
