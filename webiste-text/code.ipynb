```python
import requests
from bs4 import BeautifulSoup
import ssl



url = "https://developer.mozilla.org/en-US/docs/Web/JavaScript/Reference/Errors/Unterminated_string_literal"

# Create SSL context
ssl_context = ssl.create_default_context()

# Disable SSL certificate verification
ssl_context.check_hostname = False
ssl_context.verify_mode = ssl.CERT_NONE
html = requests.get(url=url, verify=False, cert=ssl_context)
html = html.text
soup = BeautifulSoup(html, features="html.parser")

# kill all script and style elements
for script in soup(["script", "style"]):
    script.extract()    # rip it out

# get text
text = soup.get_text()

# break into lines and remove leading and trailing space on each
lines = (line.strip() for line in text.splitlines())
# break multi-headlines into a line each
chunks = (phrase.strip() for line in lines for phrase in line.split("  "))
# drop blank lines
text = '\n'.join(chunk for chunk in chunks if chunk)

print(text)
```

```python
import requests
from bs4 import BeautifulSoup
import ssl

url = "https://blog.tharinduhasthika.com/exploring-the-infinite-rolling-median"

# Create SSL context
ssl_context = ssl.create_default_context()

# Disable SSL certificate verification
ssl_context.check_hostname = False
ssl_context.verify_mode = ssl.CERT_NONE

# Make the request with SSL certificate verification disabled
html = requests.get(url=url, verify=True)

# Get the response content as text
html_text = html.text

# Use BeautifulSoup to parse the HTML
soup = BeautifulSoup(html_text, features="html.parser")

# Kill all script and style elements
for script in soup(["script", "style"]):
    script.extract()

# Get the text from the parsed HTML
text = soup.get_text()

# Break into lines and remove leading and trailing space on each
lines = (line.strip() for line in text.splitlines())

# Break multi-headlines into a line each
chunks = (phrase.strip() for line in lines for phrase in line.split("  "))

# Drop blank lines
text = '\n'.join(chunk for chunk in chunks if chunk)

print(text)
```
