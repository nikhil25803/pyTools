```python
import PyPDF2

# Open the PDF file
pdf_file = open('../debu.pdf', 'rb')

# Create a PDF reader object
pdf_reader = PyPDF2.PdfFileReader(pdf_file)

# Create a PDF writer object
pdf_writer = PyPDF2.PdfFileWriter()

# Iterate through each page of the PDF file
for page_num in range(pdf_reader.getNumPages()):
    # Get the page object
    page = pdf_reader.getPage(page_num)

    # Add the stamp to the page
    stamp = PyPDF2.PdfFileReader(open('../gdgKolkata.pdf', 'rb')).getPage(0)  # Open the stamp PDF
    page.mergePage(stamp)  # Merge the stamp with the page

    # Add the modified page to the writer object
    pdf_writer.addPage(page)

# Write the modified PDF to a new file
with open('stamped.pdf', 'wb') as output:
    pdf_writer.write(output)

# Close the PDF files
pdf_file.close()
```
